```python
import collections

# Question 1
# 1)
dante_text_content = open("dante.txt", "r")
dante_lines = list()
for line in dante_text_content:
    if len(line.strip()) == 0:
        continue
    dante_lines.append(line)
    print(line)

dante_text_content.close()

# 2)
print("\n")

dante_lines_clean = [line.replace("\n", "") for line in dante_lines]
print(dante_lines_clean)

# 3)
print("\n")

dante_one_line = " ".join(dante_lines_clean)
print(dante_one_line)

# 4)
print("\n")

remove_punctuation = lambda string_content: string_content.strip(",.!")
print(remove_punctuation("Hello"))
print(remove_punctuation("Hello class!"))

# 5)
print("\n")

dante_lower = [remove_punctuation(line).lower() for line in dante_lines_clean]
print(dante_lower)

# 6)
print("\n")

dante_dict = dict()

for line in dante_lines_clean:
    for word in line.split():
        dante_dict[word[0]] = word

print(dante_dict)
```

```python
import numpy as np
from ineqpy.inequality import gini

household_incomes = np.loadtxt("us_household_incomes.txt")

# Question 2
# 4)
from ineqpy.inequality import gini

ex4_gini = gini(household_incomes) * 100
print(f'The gini value of the US households is {ex4_gini:.2f}')

# 5)

# Compared to the gini indices of various other countries according to https://www.indexmundi.com/facts/indicators/SI.POV.GINI/rankings, the US ranks 46th out of 162 countries.
# My index seems to be around the same at about 41%.
# This result doesn't suprise me all too much dues to how the divide has notoriously been reported on several times over the years, especially after the pandemic.

# Question 2
# 6)
print("\n")

policy_a = household_incomes.copy()
policy_a[policy_a < 40000] += 5000

policy_b = household_incomes.copy()
policy_b[policy_b < 30000] += 7000

ex6_gini_policy_a = gini(policy_a) * 100
print(f'The gini index when applying policy a is now {ex6_gini_policy_a:.2f}')

ex6_gini_policy_b = gini(policy_b) * 100
print(f'The gini index when applying policy b is now {ex6_gini_policy_b:.2f}')

results = { "ex6_gini_which_reduced_more": "Policy A" if ex6_gini_policy_a > ex6_gini_policy_b else "Policy B" }
print(f'The policy that lowered inequity more was {results["ex6_gini_which_reduced_more"]}')

# 7)
print("\n")

policy_c = household_incomes.copy()
policy_c[policy_c > 250000] *= .95
ex7_gini_policy_c = gini(policy_c) * 100
print(f'The gini index when applying policy c is now {ex7_gini_policy_c:.2f}')

results["ex7_gini_policy_c"] = ex7_gini_policy_c

# 8)
print("\n")

ex8_revenue_raised = (household_incomes - policy_c).sum()
print(f'The total revenue gained from the tax in policy C would be ${ex8_revenue_raised:,.2f}')

results["ex8_revenue_raised"] = ex8_revenue_raised

# 9)
print("\n")

num_households_less_criterion = len(household_incomes[household_incomes < 30000])
ex9_transfers = ex8_revenue_raised / num_households_less_criterion
print(f'The revenue each of the {num_households_less_criterion:,} homes would be ${ex9_transfers:,.2f}')

results["ex9_transfers"] = ex9_transfers

# 10)
print("\n")

policy_d = policy_c.copy()
policy_d[policy_d < 30000] += ex9_transfers
ex10_gini_policy_d = gini(policy_d) * 100
print(f'The gini index when applying policy d is now {ex10_gini_policy_d:.2f}')

results["ex10_gini_policy_d"] = ex10_gini_policy_d

# 11)
print("\n")

num_households_less_criterion_e = len(household_incomes[household_incomes < 40000])
ex11_transfers = ex8_revenue_raised / num_households_less_criterion_e

policy_e = policy_c.copy()
policy_e[policy_e < 40000] += ex11_transfers
ex11_gini_policy_e = gini(policy_e) * 100
print(f'The gini index when applying policy e is now {ex11_gini_policy_e:.2f}')

results["ex11_gini_policy_e"] = ex11_gini_policy_e

# 12)
print("\n")

results["ex12_policy_recommendation"] = "Policy D" if ex10_gini_policy_d > ex11_gini_policy_e else "Policy E"
print(f'The policy recommendation for distributing revenue from taxation on homes earning > $250,000 is {results["ex12_policy_recommendation"]}')

print(results)
```

```python
import pandas as pd
from ineqpy.inequality import gini

gdppercap = pd.Series([34605, 34493, 12393, 44200, 10041, 58138, 4709, 49284, 10109, 42536], index=["Bahrain", "Belgium", "Bulgaria", "Ireland", "Macedonia", "Norway", "Paraguay", "Singapore", "South Africa", "Switzerland"])

# Question 3
# 3)
ex3_highest_gpd_per_capita = gdppercap.index.max()
print(f'The country with the highest GDP/capita is {ex3_highest_gpd_per_capita}')

ex3_lowest_gpd_per_capita = gdppercap.index.min()
print(f'The country with the highest GDP/capita is {ex3_lowest_gpd_per_capita}')

# 4)
print("\n")

ex4_lessthan_20000 = gdppercap[gdppercap < 20000].index.sort_values()
print(f'The sorted list of countries w/GDP/capita < $20,000 is {ex4_lessthan_20000}')

results = { "ex4_lessthan_20000": ex4_lessthan_20000 }

# 5)
print("\n")

ex5_switzerland = gdppercap["Switzerland"]
print(f'The GDP/capita of Switzerland is ${ex5_switzerland:,.2f}')

results["ex5_switzerland"] = ex5_switzerland

# 6)
print("\n")

def gini_func(income_series):
    income_numerator = 0.0
    income_denomenator = 0.0
    num_series_elements = len(income_series)
    
    for index, income_value in enumerate(income_series):
        income_numerator += (index + 1) * income_value
        income_denomenator += income_value

    return (2 * income_numerator) / (num_series_elements * income_denomenator) - (num_series_elements + 1) / num_series_elements

results["ex6_gini_loop"] = gini_func(gdppercap)
# print(results["ex6_gini_loop"])
# print(gini(gdppercap))

# 7)
def gini_vectorized_func(income_series):
    return (2 * (range(1, len(income_series) + 1) * income_series).sum()) / (len(income_series) * income_series.sum()) - (len(income_series) + 1) / len(income_series)

results["ex7_gini_vectorized"] = gini_func(gdppercap)

# 8)
print("\n")

avg_growth = pd.Series([-0.29768835, 0.980299584, 4.52991925, 3.686556736, 2.621416804, 0.775132075, 2.015489468, 3.345793635, 1.349993318, 0.982775018], index=["Bahrain", "Belgium", "Bulgaria", "Ireland", "Macedonia", "Norway", "Paraguay", "Singapore", "South Africa", "Switzerland"])
results["ex8_gini_2025"] = gini_vectorized_func(gdppercap * (1 + (avg_growth / 100))**17)

# 9)
print(results)

# It seems that the GDP doesn't converge since it actually increased. 
```
